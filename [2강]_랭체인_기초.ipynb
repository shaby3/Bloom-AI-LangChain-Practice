# Text Generation

```python
import dotenv
dotenv.load_dotenv()
from langchain.chat_models import init_chat_model
```

```python
model = init_chat_model("gpt-4o")
```

### invoke()

```python
response = model.invoke("안녕하세요. 당신은 누구입니까?")
```

```python
response
```

```python
response.content
```

```python
response.usage_metadata
```

### parameter

```python
model = init_chat_model(
    "gpt-4o",
    temperature = 0.7,
    timeout = 5,
    max_tokens = 1000
)
```

```python
response = model.invoke("안녕하세요. 당신은 누구입니까?")
```

```python
response
```

### stream()

```python
for chunk in model.stream("AI Agent란 무엇인가요?"):
    print(chunk.text, end = '')
```

### batch()

```python
inputs = [
    "과적합이 뭔가요?",
    "앵무새의 털 색상이 화려한 이유는?",
    "AI Agent 기반 서비스는 뭐가 있나요?"
]
```

```python
responses = model.batch(inputs)

for response in responses:
    print(response.content)
    print('\n')
```

# Structured Outputs

### Pydantic

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    """상세한 영화 정보."""
    title: str = Field(description="영화의 제목")
    year: int = Field(description="개봉 연도")
    director: str = Field(description="영화 감독 이름")
    rating: float = Field(description="영화 평점 (10점 만점)")
```

```python
model_with_parsing = model.with_structured_output(Movie)
```

```python
response = model_with_parsing.invoke("영화 인터스텔라에 대해 설명해주세요.")
```

```python
print(response)
```

```python
model.invoke("영화 인터스텔라에 대해 설명해주세요.")
```

### JSON output parser

```python
import json

json_schema = {
    "title": "Movie",
    "description": "A movie with details",
    "type": "object",
    "properties": {
        "title": {
            "type": "string",
            "description": "The title of the movie"
        },
        "year": {
            "type": "integer",
            "description": "The year the movie was released"
        },
        "director": {
            "type": "string",
            "description": "The director of the movie"
        },
        "rating": {
            "type": "number",
            "description": "The movie's rating out of 10"
        }
    },
    "required": ["title", "year", "director", "rating"]
}
```

```python
model_with_json = model.with_structured_output(json_schema)
```

```python
model_with_json.invoke("영화 인터스텔라에 대해 설명해주세요.")
```

# Memory

### 방법 1

```python
from langchain.messages import HumanMessage, SystemMessage, AIMessage

system_msg = SystemMessage("당신은 유능한 로켓 전문가입니다.")
human_msg = HumanMessage("안녕하세요. 궁금한 것이 있어요!")
```

```python
messages = [system_msg, human_msg]

response = model.invoke(messages)
```

```python
response.content
```

```python
hm_1 = HumanMessage("안녕하세요. 저는 수호입니다")
aim_1 = AIMessage("안녕하세요 수호! 어떤 도움이 필요하시죠?")
hm_2 = HumanMessage("제 이름이 뭐라고 했죠?")

messages = [hm_1, aim_1, hm_2]
response = model.invoke(messages)
```

```python
response.content
```
